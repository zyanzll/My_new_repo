```python
import pandas as pd
import plotly.express as px

# Leer los datos
car_data = pd.read_csv('../vehicles_us.csv')

# Explorar la estructura básica de los datos
print("Primeras 5 filas:")
print(car_data.head())

print("\nInformación del dataset:")
print(car_data.info())

# Ver las dimensiones del dataset
print(f"El dataset tiene {car_data.shape[0]} filas y {car_data.shape[1]} columnas")

# Ver las columnas disponibles
print("\nColumnas del dataset:")
print(car_data.columns.tolist())
# Verificar valores faltantes
print("Valores faltantes por columna:")
print(car_data.isnull().sum())
# Estadísticas básicas de las columnas numéricas
print("Estadísticas descriptivas:")
print(car_data.describe())

```
